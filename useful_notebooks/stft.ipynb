```python
import pandas as pd
prices = pd.read_csv('../cryptoCompareData/btc_prices.csv')
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(ncols=2,figsize=(20,5))
spectrum, freqs, t, im  = ax1.specgram(prices['close'], NFFT=256, cmap='jet_r')
fig.colorbar(im, ax=ax1)
ax1.set_ylabel("Frequency")
ax1.set_xlabel("Time")

_, _, _, im  = ax2.specgram(np.log(prices['close']), NFFT=256, cmap='jet_r')
fig.colorbar(im, ax=ax2)
ax2.set_ylabel("Frequency")
ax2.set_xlabel("Time")
```

```python
from scipy.signal import stft

f,t,Zxx = stft(prices['close'].values,nperseg=64)
plt.pcolormesh(t, f, np.abs(Zxx))
```

```python
f,t,Zxx = stft(np.log(prices['close']).values,nperseg=64)
plt.pcolormesh(t, f, np.abs(Zxx)**2)
plt.colorbar()
```
